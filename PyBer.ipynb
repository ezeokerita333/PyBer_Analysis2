```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the DataFrame
pyber_data_df.head()
```

```python
# Create the Urban city DataFrame.
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
urban_cities_df.head()
```

```python
# Create the Suburban city DataFrame.
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
suburban_cities_df.head()
```

```python
# Create the Rural city DataFrame.
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
rural_cities_df.head()
```

```python
# Get the number of rides for urban cities.
urban_cities_count_df = urban_cities_df.groupby("city" ).count()["ride_id"]
urban_cities_count_df.head()
```

```python
suburban_cities_count_df = suburban_cities_df.groupby("city" ).count()["ride_id"]
suburban_cities_count_df.head()
```

```python
rural_cities_count_df = rural_cities_df.groupby("city" ).count()["ride_id"]
rural_cities_count_df.head()
```

```python
# Get average fare for each city in the urban cities.
urban_avg_fare = urban_cities_df.groupby(["city"]).mean()["fare"]
urban_avg_fare.head()
```

```python
# Get average fare for each city in the suburban cities.
suburban_avg_fare = suburban_cities_df.groupby(["city"]).mean()["fare"]
suburban_avg_fare.head()
```

```python
# Get average fare for each city in the rural cities.
rural_avg_fare = rural_cities_df.groupby(["city"]).mean()["fare"]
rural_avg_fare.head()
```

```python
# Get the average number of drivers for each urban city.
urban_driver_count = urban_cities_df.groupby(["city"]).mean()["driver_count"]
urban_driver_count.head()
```

```python
# Get the average number of drivers for each suburban city.
suburban_driver_count = suburban_cities_df.groupby(["city"]).mean()["driver_count"]
suburban_driver_count.head()
```

```python
# Get the average number of drivers for each rural city.
rural_driver_count = rural_cities_df.groupby(["city"]).mean()["driver_count"]
rural_driver_count.head()
```

```python
# Build the scatter plots for urban cities.
#Urban_plots = plt.scatter(urban_cities_count_df, urban_avg_fare, s=urban_driver_count)
# Build the scatter plots for urban cities.
plt.scatter(urban_cities_count_df,
      urban_avg_fare,
      s=10*urban_driver_count, c="coral",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Urban")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
# Add the legend.
plt.legend()
```

```python
# Build the scatter plots for suburban cities.
plt.scatter(suburban_cities_count_df,
      suburban_avg_fare,
      s=10*suburban_driver_count, c="skyblue",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Suburban")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
# Add the legend.
plt.legend()
#The chart in the output c
```

```python
# Build the scatter plots for rural cities.
plt.scatter(rural_cities_count_df,
      rural_avg_fare,
      s=10*rural_driver_count, c="gold",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Rural")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
# Add the legend.
plt.legend()
```

```python
# Add the scatter charts for each type of city.
plt.subplots(figsize=(10, 6))
plt.scatter(urban_cities_count_df,
      urban_avg_fare,
      s=10*urban_driver_count, c="coral",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Urban")

plt.scatter(suburban_cities_count_df,
      suburban_avg_fare,
      s=10*suburban_driver_count, c="skyblue",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Suburban")

plt.scatter(rural_cities_count_df,
      rural_avg_fare,
      s=10*rural_driver_count, c="gold",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Rural")

# Incorporate the other graph properties
plt.title("PyBer Ride-Sharing Data (2019)", fontsize=20)
plt.ylabel("Average Fare ($)", fontsize=12)
plt.xlabel("Total Number of Rides (Per City)", fontsize=12)
plt.grid(True)

# Add the legend.
plt.legend()
# Incorporate a text label about circle size.
plt.text(42, 35, "Note: Circle size correlates with driver count per city.", fontsize="12")
# Save the figure.
plt.savefig("analysis/Fig1.png")

# Show the plot
plt.show()
```

```python
# Import NumPy and the stats module from SciPy.
import numpy as np
import scipy.stats as sts
```

```python
# Calculate the measures of central tendency for the ride count for the urban cities.
mean_urban_cities_count_df = np.mean(urban_cities_count_df)
print(f"The mean for the ride counts for urban trips is {mean_urban_cities_count_df:.2f}.")

median_urban_cities_count_df = np.median(urban_cities_count_df)
print(f"The median for the ride counts for urban trips is {median_urban_cities_count_df}.")

mode_urban_cities_count_df = sts.mode(urban_cities_count_df)
print(f"The mode for the ride counts for urban trips is {mode_urban_cities_count_df}.")
```
